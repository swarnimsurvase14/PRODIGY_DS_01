#  Total Population Analysis


```python
# Filter data for total population
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']

# Sort data based on the total population for 2022
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)

# Get the top ten countries with the highest total population for 2022
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country Name']] )
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract relevant data for plotting
top_countries = total_top_ten_countries[['Country Name', '2022']]
top_countries = top_countries.rename(columns={'2022': 'Population'})

# Set plot style
sns.set(style="whitegrid")

# --------- Bar Plot ---------
plt.figure(figsize=(12, 6))
sns.barplot(x='Population', y='Country Name', data=top_countries, palette='viridis')
plt.title('Top 10 Countries by Total Population in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract relevant data for plotting
top_countries = total_top_ten_countries[['Country Name', '2015']]
top_countries = top_countries.rename(columns={'2015': 'Population'})

# Set plot style
sns.set(style="whitegrid")

# --------- Bar Plot ---------
plt.figure(figsize=(12, 6))
sns.barplot(x='Population', y='Country Name', data=top_countries, palette='viridis')
plt.title('Top 10 Countries by Total Population in 2015')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Prepare data for grouped bar plot
comparison_data = total_top_ten_countries[['Country Name', '2001', '2010']].copy()
comparison_data = comparison_data.melt(id_vars='Country Name', var_name='Year', value_name='Population')

# Plot grouped bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='Population', hue='Year', data=comparison_data, palette='Set2')
plt.title('Comparison of Total Population in 2001 vs 2010 (Top 10 Countries)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Get bottom 10 countries by population in 2022
total_bottom_ten_countries = total_population_sorted.tail(10).reset_index(drop=True)
print("Bottom ten countries by total population in 2022:\n")
print(total_bottom_ten_countries[['Country Name', '2022']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=total_bottom_ten_countries, palette='magma')

plt.title('Bottom 10 Countries by Total Population in 2022', fontsize=16)
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Select relevant columns for bottom 10 countries comparison
bottom_comparison = total_bottom_ten_countries[['Country Name', '2001', '2010']].copy()

# Melt the dataframe to long format for easier plotting
bottom_comparison_melted = bottom_comparison.melt(
    id_vars='Country Name', var_name='Year', value_name='Population')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='Population', hue='Year', data=bottom_comparison_melted, palette='Set2')

plt.title('Population Comparison of Bottom 10 Countries: 2001 vs 2010', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#  Gender-wise Population Analysis


```python
# Filter data for male population
male_population_data = df[df["Series Code"] == "SP.POP.TOTL.MA.IN"]

# Sort data based on the male population for 2022
male_population_sorted = male_population_data.sort_values(by="2022", ascending=False)

# Get the top ten countries with the highest male population for 2022
male_top_ten_countries = male_population_sorted.head(10)
print("Top ten countries of male population")
print(male_top_ten_countries[['Country Name']] )
```

```python
# Filter data for female population
female_population_data = df[df["Series Code"] == "SP.POP.TOTL.FE.IN"]

# Sort data based on the female population for 2022
female_population_sorted = female_population_data.sort_values(by="2022", ascending=False)

# Get the top ten countries with the highest female population for 2022
female_top_ten_countries = female_population_sorted.head(10)
print("Top ten countries of female population")
print(female_top_ten_countries[['Country Name']] )
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for male population
plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=male_top_ten_countries, palette='Blues_r')

plt.title('Top 10 Countries by Male Population (2022)', fontsize=16)
plt.xlabel('Male Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

# Bar plot for female population
plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=female_top_ten_countries, palette='Reds_r')

plt.title('Top 10 Countries by Female Population (2022)', fontsize=16)
plt.xlabel('Female Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
male_bottom_ten_countries = male_population_sorted.tail(10)
print("Top ten countries of lowest male population")
print(male_bottom_ten_countries[['Country Name']] )
```

```python
female_bottom_ten_countries = female_population_sorted.tail(10)
print("Top ten countries of lowest female population")
print(female_bottom_ten_countries[['Country Name']] )
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=male_bottom_ten_countries, palette='Blues')

plt.title('Bottom 10 Countries by Male Population (2022)', fontsize=16)
plt.xlabel('Male Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=female_bottom_ten_countries, palette='Reds')

plt.title('Bottom 10 Countries by Female Population (2022)', fontsize=16)
plt.xlabel('Female Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Get bottom 10 country names based on male/female population
bottom_countries = male_bottom_ten_countries['Country Name'].tolist()

# Filter both male and female data for those countries
male_bottom = male_population_data[male_population_data['Country Name'].isin(bottom_countries)]
female_bottom = female_population_data[female_population_data['Country Name'].isin(bottom_countries)]

# Create a DataFrame for combined gender comparison
gender_comparison_bottom = pd.DataFrame({
    'Country Name': bottom_countries,
    'Male Population 2022': male_bottom.set_index('Country Name')['2022'],
    'Female Population 2022': female_bottom.set_index('Country Name')['2022']
}).reset_index(drop=True)

# Melt the data for seaborn
gender_comparison_melted = gender_comparison_bottom.melt(
    id_vars='Country Name',
    value_vars=['Male Population 2022', 'Female Population 2022'],
    var_name='Gender',
    value_name='Population'
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='Population', hue='Gender', data=gender_comparison_melted, palette='Set2')

plt.title('Male vs Female Population (2022) — Bottom 10 Countries', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Get top 10 country names based on male/female population
top_countries = male_top_ten_countries['Country Name'].tolist()

# Filter both male and female data for those countries
male_top = male_population_data[male_population_data['Country Name'].isin(top_countries)]
female_top = female_population_data[female_population_data['Country Name'].isin(top_countries)]

# Create combined DataFrame for gender comparison
gender_comparison_top = pd.DataFrame({
    'Country Name': top_countries,
    'Male Population 2022': male_top.set_index('Country Name')['2022'],
    'Female Population 2022': female_top.set_index('Country Name')['2022']
}).reset_index(drop=True)

# Melt for plotting
gender_comparison_top_melted = gender_comparison_top.melt(
    id_vars='Country Name',
    value_vars=['Male Population 2022', 'Female Population 2022'],
    var_name='Gender',
    value_name='Population'
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='Population', hue='Gender', data=gender_comparison_top_melted, palette='Set1')

plt.title('Male vs Female Population (2022) — Top 10 Countries', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot setup
plt.figure(figsize=(12, 6))
bar1 = plt.bar(stacked_df['Country'], stacked_df['Male'], label='Male', color='skyblue')
bar2 = plt.bar(stacked_df['Country'], stacked_df['Female'], bottom=stacked_df['Male'], label='Female', color='lightcoral')

# Labels and title
plt.title('Stacked Bar Plot: Male vs Female Population (2022) — Top 10 Countries', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
